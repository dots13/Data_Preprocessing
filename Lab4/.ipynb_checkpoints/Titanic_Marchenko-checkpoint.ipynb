```python
%matplotlib inline
import numpy as np
import pandas as pd
import sklearn as skit
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.style.use('ggplot')
import re
```

Load data from the csv and prepare them:

      - Data (columns) not present in data1.csv are omitted from data2.csv.
      - Calculate the Age using the BirthYear column in data2.csv and append the result to the Age column of data1.
      - PasangerId must be unique in the resulting data frame. 
      - Use pandas.concat method

```python
data1 = pd.read_csv('data1.csv')
data2 = pd.read_csv('data2.csv',sep=';')
df1 = data1
df2 = data2
df2['Age'] = 1912 -(df2['BirthYear'])
data = pd.concat([df1, df2], join = 'inner', ignore_index = True)
data['PassengerId'] = range(1,len(data)+1)
```

From column "Name" extract 'titles'

```python
def get_title(name):
    # Use a regular expression to search for a title.  Titles always consist of capital and lowercase letters, and end with a period.
    title_search = re.search(' ([A-Za-z]+)\.', name)
    # If the title exists, extract and return it.
    if title_search:
        return title_search.group(1)

# Get all the titles and print how often each one occurs.
data["titles"] = data["Name"].apply(get_title)
```

```python
titles_list = ('Mr', 'Mrs', 'Miss', 'Master')
data = data.loc[data['titles'].isin(titles_list)]
```

plot#1 show survived people and their age in different class cabins

```python
plt.figure(figsize=(18,8))
sns.stripplot(x="Pclass", y="Age", hue="Survived", data=data, palette= ['black','red'], jitter=True)
```

plot#2 show survived people by different titles

```python
plt.figure(figsize=(18,8))
sns.stripplot(x="titles", y="Age", hue="Survived", data=data,size = 10, alpha=.75, jitter=.2, palette= ['black','red'])
```
