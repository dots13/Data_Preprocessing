```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import requests
import pandas as pd
```

```python
web_link = list()
df_list = list()
for i in range(1,15):
    web_link.append('https://volby.cz/pls/ps2017/ps111?xjazyk=CZ&xkraj=' + str(i) + '&xstrana=0&xv=1&xt=1')
```

```python
web_link
df = pd.DataFrame()
r = requests.get('https://volby.cz/pls/ps2017/ps111?xjazyk=CZ&xkraj=1&xstrana=0&xv=1&xt=1')
html = r.text
list_of_data_frames = pd.read_html(html,flavor='html5lib')
df = list_of_data_frames[0]
```

get data from all pages

```python
for url in web_link:
    r = requests.get(url)
    html = r.text
    list_of_data_frames = pd.read_html(html,flavor='html5lib')
    df_list.append(list_of_data_frames[0])
```

date cleaning

```python
clear_df = list()
for i in range(1,14):
    clear_df.append(df_list[i][[1,0,4,3]].copy())
    clear_df[i-1].rename(columns={1: 'name', 0: 'number', 4 : 'age', 3 : 'all_name'}, inplace=True)
    clear_df[i-1] = clear_df[i-1][1:]
```

```python
df_all = pd.concat(clear_df)
```

```python
popular_list = ('ČSSD', 'ANO', 'KSČM', 'TOP 09', 'ODS', 'KDU-ČSL')
df_all = df_all.loc[df_all['name'].isin(popular_list)]
```

```python
Educ_list = ('Mgr.', 'Ing.', 'Bc.', 'No Education')
```

```python
df_all['f_name'] = df_all['all_name'].str.split(' ').str.get(0)
df_all['f_name'] = df_all['all_name'].str.split(' ').str.get(1)
df_all['titul'] = df_all['all_name'].str.split(' ').str.get(2) 
df_all = df_all.replace(np.nan, 'No Education', regex=True)
df_all.titul[~df_all['titul'].isin(Educ_list)] = 'High Education'
```

```python
df_all['educ']= df_all['titul'].apply(lambda x: 0 if x == 'No Education' else 1)
```

plot#1 How many people in different parties have an education (in percent)

```python
sns.barplot(x="name", y="educ",data=df_all, palette = "husl");
```

plot#2 HHow many people in different parties have an education and will show the level of education

```python
sns.factorplot(x="name", hue="titul", data=df_all, kind="count", size=10, palette = "PuBuGn_d");
```

plot#3 Shows the number of people depending on the age with education and without education

```python
sns.factorplot(x = "age", hue="educ", data=df_all, kind="count", size=10, palette = "PuBuGn_d");
```

plot#4 Shows the percentage of people with different levels of education

```python
df_all.titul.value_counts().plot(kind='pie', subplots=True,  autopct='%1.0f%%', figsize=(8, 8))
```

plot#5 Shows how many people and with what level of education each party has (and their age)

```python
plt.figure(figsize=(30,15))
g = sns.stripplot(x="name", y="age", hue = "titul", data=df_all, alpha=.45, size = 20, jitter=0.35, dodge=True)
```

plot#6 shows party members for each level of education

```python
plt.figure(figsize=(30,15))
g = sns.stripplot(x="titul", y="age", hue = "name", data=df_all, alpha=.75, size = 20, jitter=0.35, palette="Set1", dodge=True)
```
